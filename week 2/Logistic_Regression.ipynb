```python
#Nama : Fasya Hano=ifah
#NIM : 1103200149
#Logistic Regression Purchase
import pandas as pd

# Ganti path sesuai dengan lokasi dataset CSV di Google Drive Anda
file_path = '/content/sample_data/Social_Network_Ads.csv'

# Membaca dataset
df = pd.read_csv(file_path)

# Menampilkan beberapa baris pertama dataset
print(df.head())
```

```python
df.head() #Membaca 5 baris teratas data frame
```

```python
df.info() #Menampilkan informasi mengenai DataFrame
```

```python
df= df.drop(columns=['User ID']) #Hapus kolom yang tidak penting untuk dipelajari mode;
```

```python
#One hot encoding dengan pd.get_dumnies()
df = pd.get_dummies(df)
df
```

```python
# pisahkan atribut dan label
predictions = ['Age' , 'EstimatedSalary' , 'Gender_Female' , 'Gender_Male']
X = df[predictions]
y = df['Purchased']
```

```python
# lakukan normalisasi terhadap data yang kita miliki
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns = X.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split

#bagi data menjadi train dan test set
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

```python
from sklearn import linear_model

# Membuat model logistic regression
model = linear_model.LogisticRegression()

# Latih model dengan fungsi fit()
model.fit(X_train, y_train)

# Mendapatkan parameter-parameter model
params = model.get_params()
print("LogisticRegression")
print(params)
```

```python
#Uji akurasi model
model.score(X_test, y_test)
```
